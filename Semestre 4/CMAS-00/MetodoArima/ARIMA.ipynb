```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(f_birth)
```

```python
plot_pacf(f_birth)
```

```python
f_birth.size
```

```python
birth_train = f_birth[0:330]
birth_test = f_birth[330:365]
```

```python
birth_train.size
```

```python
birth_test.size
```

```python
from statsmodels.tsa.ar_model import ARIMA
```

```python
birth_model = ARIMA(birth_train, order=(2,0,3))
```

```python
birth_model_fit = birth_model.fit()
```

```python
birth_model_fit.aic 
```

```python
birth_forecast = birth_model_fit.forecast(step=35)[0]
```

```python
birth_forecast
```

```python
birth_test
```

```python
np.sqrt(mean_squared_error(birth_test, birth_forecast))
```
